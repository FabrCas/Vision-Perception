# load images

```python
img_1 = cv2.imread("./2.jfif", 0)
img_2 = cv2.imread("./7.jfif")
use_gaussian = True

if use_gaussian:
  img_1 = cv2.GaussianBlur(img_1,(5,5),0)
  img_2 = cv2.GaussianBlur(img_2,(5,5),0)
  cv2_imshow(img_1)
  cv2_imshow(img_2)
else:
  cv2_imshow(img_1)
  cv2_imshow(img_2)

```

```python
# derivative phase

# define the strucuture of Prewitt and Roberts filters
kernel_x_prewitt = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
kernel_y_prewitt = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
kernel_robertsx = np.array([[-1, 0], [0, 1]])
kernel_robertsy = np.array([[0, -1], [1, 0]])


laplacian = cv2.Laplacian(img_1,cv2.CV_64F, borderType=cv2.BORDER_DEFAULT)
sobelx = cv2.Sobel(img_1,cv2.CV_64F,1,0,ksize=5,borderType=cv2.BORDER_DEFAULT )
sobely = cv2.Sobel(img_1,cv2.CV_64F,0,1,ksize=5, borderType=cv2.BORDER_DEFAULT)
sobel_magnitude = np.sqrt(sobelx**2 + sobely**2)
scharrx = cv2.Scharr(img_1,ddepth=-1, dx=1, dy=0, borderType=cv2.BORDER_DEFAULT)
scharry = cv2.Scharr(img_1,ddepth=-1, dx=0, dy= 1, borderType=cv2.BORDER_DEFAULT)
prewittx = cv2.filter2D(img_1, -1, kernel_x_prewitt)
prewitty = cv2.filter2D(img_1, -1, kernel_y_prewitt)
robertsx = cv2.filter2D(img_1, -1, kernel_robertsx)
robertsy = cv2.filter2D(img_1, -1, kernel_robertsy)

laplacian_2 = cv2.Laplacian(img_2,cv2.CV_64F, borderType=cv2.BORDER_DEFAULT)
sobelx_2 = cv2.Sobel(img_2,cv2.CV_64F,1,0,ksize=5, borderType=cv2.BORDER_DEFAULT)
sobely_2 = cv2.Sobel(img_2,cv2.CV_64F,0,1,ksize=5, borderType=cv2.BORDER_DEFAULT)
sobel_magnitude_2 = np.sqrt(sobelx_2**2 + sobely_2**2)
scharrx_2 = cv2.Scharr(img_2,ddepth=-1, dx=1, dy=0, borderType=cv2.BORDER_DEFAULT)
scharry_2 = cv2.Scharr(img_2,ddepth=-1, dx=0, dy= 1, borderType=cv2.BORDER_DEFAULT)
prewittx_2 = cv2.filter2D(img_2, -1, kernel_x_prewitt)
prewitty_2 = cv2.filter2D(img_2, -1, kernel_y_prewitt)
robertsx_2 = cv2.filter2D(img_2, -1, kernel_robertsx)
robertsy_2 = cv2.filter2D(img_2, -1, kernel_robertsy)

# cv2.imwrite("laplacian_1_img.jpg", laplacian)
# cv2.imwrite("robertsy_1_img.jpg", robertsy)


# cv2.imwrite("gen_img/prewittx_2_img.jpg", prewittx_2)
# cv2.imwrite("gen_img/prewitty_2_img.jpg", prewitty_2)


# showing phase

plt.imshow(img_1,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.show()
plt.imshow(laplacian,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.show()
plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.show()
plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])
plt.show()
plt.imshow(sobel_magnitude,cmap = 'gray')
plt.title('Sobel Magnitude'), plt.xticks([]), plt.yticks([])
plt.show()
plt.imshow(scharrx,cmap = 'gray')
plt.title('scharr x'), plt.xticks([]), plt.yticks([])
plt.show()
plt.imshow(scharry,cmap = 'gray')
plt.title('scharr y'), plt.xticks([]), plt.yticks([])
plt.show()
plt.imshow(prewittx,cmap = 'gray')
plt.title('Prewitt x'), plt.xticks([]), plt.yticks([])
plt.show()
plt.imshow(prewitty,cmap = 'gray')
plt.title('Prewitt y'), plt.xticks([]), plt.yticks([])
plt.show()
plt.imshow(robertsx,cmap = 'gray')
plt.title('Roberts x'), plt.xticks([]), plt.yticks([])
plt.show()
plt.imshow(robertsy,cmap = 'gray')
plt.title('Roberts y'), plt.xticks([]), plt.yticks([])
plt.show()
```
